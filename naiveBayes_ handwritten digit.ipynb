```
import numpy as np
import scipy.io
import math
import geneNewData
def calculate_probability(x, mean, var):
    exponent = math.exp(-((x-mean)**2 / (2 * var )))
    return (1 / (math.sqrt(2 * math.pi) * math.sqrt(var))) * exponent
def NBclassifier(digit0, digit1):
    if(digit0>digit1):
        return 0
    else:
        return 1

def main():
    myID='0900'
    geneNewData.geneData(myID)
    Numpyfile0 = scipy.io.loadmat('digit0_stu_train'+myID+'.mat')
    Numpyfile1 = scipy.io.loadmat('digit1_stu_train'+myID+'.mat')
    Numpyfile2 = scipy.io.loadmat('digit0_testset'+'.mat')
    Numpyfile3 = scipy.io.loadmat('digit1_testset'+'.mat')
    train0 = Numpyfile0.get('target_img')
    train1 = Numpyfile1.get('target_img')
    test0 = Numpyfile2.get('target_img')
    test1 = Numpyfile3.get('target_img')
    print([len(train0),len(train1),len(test0),len(test1)])
    print('Your trainset and testset are generated successfully!')
    #TASK1-training0
    train_0 =[]
    train_1 =[]
    for image in train0:
        mean = np.mean(image)
        std = np.std(image)
        train_0.append([mean,std])
    train_0=np.array(train_0)
    #TASK1-training1
    for image in train1:
        mean = np.mean(image)
        std = np.std(image)
        train_1.append([mean,std])
    train_1=np.array(train_1)
    test_0 =[]
    test_1 =[]
    for image in test0:
        mean = np.mean(image)
        std = np.std(image)
        test_0.append([mean,std])
    test_0=np.array(test_0)
    #TASK1-test1
    for image in test1:
        mean = np.mean(image)
        std = np.std(image)
        test_1.append([mean,std])
    test_1=np.array(test_1)

    #TASK2-(No.1) Mean of feature1 for digit0
    mean_feature1_digit0 = np.mean(train_0,axis=0)
    print("Mean of feature1 for digit0",mean_feature1_digit0[0])


    #TASK2-(No.2) Variance of feature1 for digit0
    variance_feature1_digit0 = np.var(train_0,axis=0)
    print("Variance of feature1 for digit0",variance_feature1_digit0[0])

    #TASK2-(No.3) Mean of feature2 for digit0
    mean_feature2_digit0 = np.mean(train_0,axis=0)
    print("Mean of feature2 for digit0",mean_feature2_digit0[1])

    #TASK2-(No.4) Variance of feature2 for digit0
    variance_feature2_digit0 = np.var(train_0,axis=0)
    print("Variance of feature2 for digit0",variance_feature2_digit0[1])

    #TASK2-(No.5) Mean of feature1 for digit1
    mean_feature1_digit1 = np.mean(train_1,axis=0)
    print("Mean of feature1 for digit1",mean_feature1_digit1[0])

    #TASK2-(No.6) Variance of feature1 for digit1
    variance_feature1_digit1 = np.var(train_1,axis=0)
    print("Variance of feature1 for digit1",variance_feature1_digit1[0])

    #TASK2-(No.7) Mean of feature2 for digit1
    mean_feature2_digit1 = np.mean(train_1,axis=0)
    print("Mean of feature2 for digit1",mean_feature2_digit1[1])

    #TASK2-(No.8) Variance of feature2 for digit1
    variance_feature2_digit1 = np.var(train_1,axis=0)
    print("Variance of feature2 for digit1",variance_feature2_digit1[1])
    countCorrect_digit0=0
    countCorrect_digit1=0
    for image in test_0:
        # p(digit0|feature1,feature2) =p(y=0)* p(feature1|digit0)*p(feature2|digit0)*p(digit0)
        digit0_feature1_feature2 = 0.5*calculate_probability(image[0],mean_feature1_digit0[0],variance_feature1_digit0[0])*calculate_probability(image[1],mean_feature2_digit0[1],variance_feature2_digit0[1])

        # p(digit1|feature1,feature2) = p(y=1)*p(feature1|digit1)*p(feature2|digit1)*p(digit1)
        digit1_feature1_feature2 = 0.5*calculate_probability(image[0],mean_feature1_digit1[0],variance_feature1_digit1[0])*calculate_probability(image[1],mean_feature2_digit1[1],variance_feature2_digit1[1])
        
        if(NBclassifier(digit0_feature1_feature2,digit1_feature1_feature2)==0):
            countCorrect_digit0+=1
    print("digit0 prediction accuracy is ",countCorrect_digit0/980)
    for image in test_1:
        # p(digit0|feature1,feature2) =p(y=0)* p(feature1|digit0)*p(feature2|digit0)*p(digit0)
        digit0_feature1_feature2 = 0.5*calculate_probability(image[0],mean_feature1_digit0[0],variance_feature1_digit0[0])*calculate_probability(image[1],mean_feature2_digit0[1],variance_feature2_digit0[1])

        # p(digit1|feature1,feature2) = p(y=1)*p(feature1|digit1)*p(feature2|digit1)*p(digit1)
        digit1_feature1_feature2 = 0.5*calculate_probability(image[0],mean_feature1_digit1[0],variance_feature1_digit1[0])*calculate_probability(image[1],mean_feature2_digit1[1],variance_feature2_digit1[1])
        
        if(NBclassifier(digit0_feature1_feature2,digit1_feature1_feature2)==1):
            countCorrect_digit1+=1
    print("digit0 prediction accuracy is ",countCorrect_digit1/1135)
if __name__ == '__main__':
    main()
```
